```python
#8 暗号
import re
alphaReg = re.compile(r'^[a-z]+$')

def isloweralpha(s):
    return alphaReg.match(s) is not None

def cipher(txt):
    l = list(txt)
    i = 0
    for c in l:
        if isloweralpha(c):
            c = chr(219-ord(c))
            l[i] = c
        i += 1
    return str(l)

cipher("Hello, I'm hogehoge.")
```

```python
#9
import random

org_s9 = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
s9list = org_s9.split()
ans9 = []

for s9 in s9list:
    if len(s9) <= 4:
        pass
    else:
        charlist = list(s9)
        midlist = charlist[1:-1]
        random.shuffle(midlist)
        s9 = s9[0] + "".join(midlist) + s9[-1]
    ans9.append(s9)

ans9 = " ".join(ans9)
ans9
```

```python
#10
def process_MeCab(mecab_filename):

    with open(mecab_filename, encoding='utf-8') as mecab_file:
        sentense = []
        sentenses = []

        for morpheme in mecab_file.read().split('\n'):
            # 　→　表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音
            surfase = morpheme.split('\t')
            if len(surfase) >= 2:            # EOS対象外
                # カンマで区切る
                result = surfase[1].split(',')
                # 結果を格納
                word = {
                    'surface' : surfase[0],
                    'base' : result[6],
                    'pos' : result[0],
                    'pos1' : result[1],
                    #'orgform' : result[5]
                }
                # 文単位に追加
                sentense.append(word)
                # 句点から終了
                if word['pos1'] == '句点':
                    sentenses.append(sentense)
                    sentense = []

    return sentenses

sentences = process_MeCab(r'C:\Users\171744\Downloads\DAstudy\0625-0629_Python（Hackathon）\02.NLP\neko.txt.mecab.txt')
sentences[:10]
```
