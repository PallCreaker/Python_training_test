```
#Set Package
import pandas as pd
import numpy as np
from bokeh.models.annotations import Title
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```
#Load CSV
df = pd.read_csv("NHANES_sample_data.csv")
del_cols = ["RIDRETH3","WTINT2YR","WTMEC2YR",
            "DR1TCARB","DR1TTFAT","LBXGLU",
            "BMXHT"]
df = df.drop(del_cols, axis=1)
df.iloc[0:1,:]
df.columns = ["SEQN","Age","Gender","Body_weight","BMI","HbA1c","Glucose"]
```

```
#Visualization
#Q1----------
df_male = df[df["Gender"]==1]
output_notebook()
p = figure(title = "HbA1c - Male", x_axis_label = "Age", y_axis_label = "HbA1c")
p.circle(df_male["Age"], df_male["HbA1c"], radius=0.5, fill_color="purple", fill_alpha=0.6, line_color=None)
show(p)
```

```
#Q2----------
df_female = df[df["Gender"]==2]
output_notebook()
p = figure(title = "BMI - Female", x_axis_label = "Age", y_axis_label = "BMI")
p.circle(df_male["Age"], df_male["BMI"], radius=0.5, fill_color=None, fill_alpha=0.6, line_color="pink")
show(p)
```
