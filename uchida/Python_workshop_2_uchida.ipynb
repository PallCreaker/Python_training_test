```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("NHANES_sample_data.csv")
del df["RIDRETH3"], df["WTINT2YR"], df["WTMEC2YR"], df["DR1TCARB"]
del df["DR1TTFAT"], df["LBXGLU"], df["BMXHT"]
df.columns = ["ID", "Age", "Gender", "Body_weight", "BMI", "HbA1c", "Glucose"]
df = df.dropna()
```

```python
# Q1
p = figure(plot_width = 600, plot_height = 600,
          title = 'HbA1c - Male',
          x_axis_label='Male', y_axis_label='HbA1c')

p.circle(df[df["Gender"]==1].loc[:, "Age"]
         , df[df["Gender"]==1].loc[:, "HbA1c"], color="purple")
output_notebook()
show(p)
```

```python
# Q2
p = figure(plot_width = 600, plot_height = 600,
          title = 'BMI - Male',
          x_axis_label='Male', y_axis_label='BMI')

p.circle_cross(df[df["Gender"]==2].loc[:, "Age"]
         , df[df["Gender"]==2].loc[:, "BMI"], color="pink")
output_notebook()
show(p)
```
