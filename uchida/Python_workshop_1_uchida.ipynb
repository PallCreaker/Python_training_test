# Pandas

```python
# Q1
df = pd.read_csv("NHANES_sample_data.csv")
df.iloc[0:30, :]
```

```python
# Q2
del df["RIDRETH3"], df["WTINT2YR"], df["WTMEC2YR"], df["DR1TCARB"]
del df["DR1TTFAT"], df["LBXGLU"], df["BMXHT"]
df.columns
```

```python
# Q3
df.columns = ["ID", "Age", "Gender", "Body_weight", "BMI", "HbA1c", "Glucose"]
df.columns
```

```python
# Q4
len(df.index)
df = df.dropna()
len(df.index)
```

```python
# Q5
df["Height"] = np.sqrt(df["Body_weight"]/df["BMI"]).round(decimals=2)
df["Height"].head()
```

```python
# Q6
pd.concat([df.median(), df.mean()], axis=1, keys=["median", "mean"])
```

```python
# Q7
pd.concat([df.mean(), df.min(), df.max()
            , df.quantile(.25), df.quantile(.75), df.std()], axis=1
            , keys=["mean", "min", "max", "quantile25", "quantile75", "std"])
```

```python
# Q8

# 条件式
age_if = df["Age"]>=20
#bmi_if = 18.0<=df["BMI"]<=50 ...これできない
bmi_if = (18.0<=df["BMI"]) & (df["BMI"]<=50)
Glucose_if = 4.0<=df["Glucose"]
HbA1c_if = (4.0<=df["HbA1c"]) & (df["HbA1c"]<=10)
df_lim_if = age_if & bmi_if & Glucose_if & HbA1c_if

# filter
df_lim = df[df_lim_if]
len(df_lim)
```

```python
# Q13

age1_if = (20<=df_lim["Age"]) & (df_lim["Age"]<40)
age2_if = (40<=df_lim["Age"]) & (df_lim["Age"]<60)
age3_if = df_lim["Age"]>=60
male_if = df_lim["Gender"]==1
female_if = df_lim["Gender"]==2
subgroup_dict = {
    "g1": df_lim[age1_if & male_if],
    "g2": df_lim[age1_if & (female_if)],
    "g3": df_lim[age2_if & male_if],
    "g4": df_lim[age2_if & (female_if)],
    "g5": df_lim[age3_if & male_if],
    "g6": df_lim[age3_if & (female_if)]
}

subgroup_dict["g6"]
```

```python
# Q13 
df_lim_wg = df_lim
age1_if = (20<=df_lim_wg["Age"]) & (df_lim_wg["Age"]<40)
age2_if = (40<=df_lim_wg["Age"]) & (df_lim_wg["Age"]<60)
age3_if = df_lim_wg["Age"]>=60
male_if = df_lim_wg["Gender"]==1
female_if = df_lim_wg["Gender"]==2

# bin作成
df_age_g = pd.cut(df_lim_wg["Age"],[19, 39, 59, df_lim_wg["Age"].max()]
       , labels=["Age1", "Age2", "Age3"])
df_age_g.columns = ["Age_label"]
df_gen_g = pd.cut(df_lim_wg["Gender"],[0, 1, 2], labels=["male", "female"])
df_gen_g.columns = ["Gen_label"]

#カラム名
key_label = [i for i in df_lim_wg.columns]
key_label.append("Age_label")
key_label.append("Gen_label")
df_lim_wg = pd.concat([df_lim_wg, df_age_g, df_gen_g], axis=1)
df_lim_wg.columns = key_label

# multiindex
df_lim_wg = df_lim_wg.set_index("Age_label")
df_lim_wg = df_lim_wg.set_index("Gen_label", append=True)
df_lim_wg.head()
subgroup_count = df_lim_wg.groupby(level=[0, 1])
subgroup_count.count()
```
