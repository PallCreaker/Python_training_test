```python
# Common Operations:1

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rename_dict = {
    "RIDAGEYR":"Age"
    ,"RIAGENDR":"Gender"
    ,"BMXWT":"Body_weight"
    ,"BMXBMI":"BMI"
    ,"LBXGH":"HbA1c"
    ,"LBDGLUSI":"Glucose"
    }
df = pd.read_csv("C:/Users/171744/NHANES_sample_data.csv").rename(columns=rename_dict).dropna().reset_index(drop=True)
df_lim = df[(df['Age'] >= 20)
             & (df['BMI'] <= 50.0)
             & (df['BMI'] >= 18.0)
             & (df['Glucose'] >= 4)
             & (df['Glucose'] <= 16)
             & (df['HbA1c'] <= 10)
             & (df['HbA1c'] >= 4)
           ]
df_lim_dropped = df_lim.drop("Gender",axis=1)
corr = df_lim_dropped.corr()
print(corr)
c = corr.abs()
c = c.unstack()
so = c.sort_values(kind="quicksort")
print(so)
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import sys

# check size
def check_min_val(min_val, euc_dist):
    if min_val <= euc_dist:
        return False
    elif min_val > euc_dist:
        return True
    else:
        return False

np_df = np.array(df_lim[["BMI","Glucose"]])
scaled_np_df = preprocessing.scale(np_df)
km = KMeans(n_clusters=4).fit(scaled_np_df)
km.cluster_centers_

y_kmeans = km.predict(scaled_np_df)
print("y_kmeans")
print(y_kmeans)
# BMI
print(scaled_np_df[:, 0])
# Glucose
print(scaled_np_df[:, 1])
# s is size of samples.
plt.scatter(scaled_np_df[:, 0], scaled_np_df[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = km.cluster_centers_
# s is size of a circle of centroid.
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("BMI")
plt.ylabel("Glucose")
# Find least euclidean distance 
nearest_dict = {}
for i in range(4):
    clust = centers[i]
    min_val = sys.maxsize
    min_index = -1
    for j in range(len(scaled_np_df)):
        # vector(np) - vector(centroid)
        euc_dist = np.linalg.norm(scaled_np_df[j,:] - clust)
        # if true, update min_val and min_index
        if check_min_val(min_val, euc_dist):
            min_val = euc_dist
            min_index = j
    # add to dict
    nearest_dict[i]=min_index
    
print("Clusters are:")
print("Cluster1:"+ str(centers[0]))
print("Cluster2:"+ str(centers[1]))
print("Cluster3:"+ str(centers[2]))
print("Cluster4:"+ str(centers[3]))

print("Nearest Vals are:")
for key, value in nearest_dict.items():
    print("Cluster" + str(key) + ":" + str(np_df[value]))
```
